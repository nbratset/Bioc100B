```python
import pandas as pd
import matplotlib.pyplot as plt

'''
<RLU> vs log antibody concentration for polyclonal anti-spike data
'''

linStyle= "--"
tablePoly = pd.read_csv("polyclonal anti-spike -Table 2.csv")
Poly = plt.figure() 

# MGATgD pre-fusion trimer
plt.errorbar("log[ug/mL]","avg[MGATgD pre-fusion trimer]", yerr=tablePoly["std[MGATgD pre-fusion trimer]"], data=tablePoly, linestyle=linStyle, label="error[MGATgD pre-fusion trimer]", capsize=3)
plt.scatter("log[ug/mL]","avg[MGATgD pre-fusion trimer]", marker=".", data=tablePoly, label="MGATgD pre-fusion trimer")

# McLellan MGAT pre-fusion trimer
plt.errorbar("log[ug/mL]","avg[McLellan MGAT pre-fusion trimer]", yerr=tablePoly["std[McLellan MGAT pre-fusion trimer]"], data=tablePoly, linestyle=linStyle, label="error[McLellan MGAT pre-fusion trimer]", capsize=3)
plt.scatter("log[ug/mL]","avg[McLellan MGAT pre-fusion trimer]", marker=".", data=tablePoly, label='McLellan MGAT pre-fusion trimer')

# McLellan CHO pre-fusion trimer
plt.errorbar("log[ug/mL]","avg[McLellan CHO pre-fusion trimer]", yerr=tablePoly["std[McLellan CHO pre-fusion trimer]"], data=tablePoly, linestyle=linStyle, label="error[McLellan CHO pre-fusion trimer]", capsize=3)
plt.scatter("log[ug/mL]","avg[McLellan CHO pre-fusion trimer]", marker=".", data=tablePoly, label='McLellan CHO pre-fusion trimer')

# HEK 293 (Genscript) S1 domain protein
plt.errorbar("log[ug/mL]","avg[HEK 293 (Genscript) S1 domain protein]", yerr=tablePoly["std[HEK 293 (Genscript) S1 domain protein]"], data=tablePoly, linestyle=linStyle, label="error[HEK 293 (Genscript) S1 domain protein]", capsize=3)
plt.scatter("log[ug/mL]","avg[HEK 293 (Genscript) S1 domain protein]", marker=".", data=tablePoly, label='HEK 293 (Genscript) S1 domain protein')

# MGAT gD-Env
plt.errorbar("log[ug/mL]","avg[MGAT gD-Env]", yerr=tablePoly["std[MGAT gD-Env]"], data=tablePoly, linestyle=linStyle, label="error[MGAT gD-Env]", capsize=3)
plt.scatter("log[ug/mL]","avg[MGAT gD-Env]", marker=".", data=tablePoly, label='MGAT gD-Env')

plt.title("Antigen Binding Curves for Polyclonal Anti-Spike")
plt.xlabel("Log of Antibody Concentration")
plt.ylabel("Relative Luminosity Units")

plt.legend(loc="upper left", bbox_to_anchor=(1.05,1), title="Key")
plt.savefig("polyclonal.jpg", dpi=600, bbox_inches="tight")
plt.show()
```
